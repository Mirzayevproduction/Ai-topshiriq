```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt


```

```python
np.random.seed(2)
X = np.linspace(0, 24, 100).reshape(-1, 1)  
y = 10 + 10 * np.sin((X.ravel() - 10) * np.pi / 29) + np.random.normal(0, 1, 100)  
```

```python

svm = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.5)
```

```python

svm.fit(X, y)
```

```python

y_pred = svm.predict(X)
```

```python

plt.figure(figsize=(10, 5))
plt.plot(X, y, label='Haqiqiy harorat', color='lightblue')
plt.plot(X, y_pred, color='blue', label='SVM Regression')
plt.xlabel("Soat")
plt.ylabel("Harorat (°C)")
plt.title("SVM Regression: Haroratni kun davomida bashorat qilish")
plt.legend()
plt.grid(True)
plt.show()
```
